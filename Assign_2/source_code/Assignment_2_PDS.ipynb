## The provided data contains various details and attributes associated with used cars. The target variable, which is the central focus of analysis, is the price of the used cars, and it is measured in lakhs. The data in this dataset is tabular, with rows and columns, where each row represents a specific used car listing, and each column represents a particular attribute or feature of these cars. Features are Make and model of the car, Location or city of sale, Year of manufacture, Mileage, Odometer(kilometers driven), Fuel type (petrol or diesel), Transmission type (manual or automatic), Number of owners, Engine displacement, Engine horsepower, Number of seats, and Price when the car was new.

```python
import pandas as pd

# Load the dataset
file_path = '/content/drive/MyDrive/train.csv'
data = pd.read_csv(file_path)

# Display the first few rows
data.head()
```

# A) Look for the missing values in all the columns and either impute them (replace with mean,median, or mode) or drop them. Justify your action for this task.

```python
# Checking for missing values
missing_values = data.isnull().sum()

# Display missing values count
missing_values
```

```python
import numpy as np
```

```python
# Importing numpy for NaN representation
import numpy as np

# replace 'null' with np.nan before conversion
data['Power'] = data['Power'].str.replace(' bhp', '').replace('null', np.nan).astype(float)

# Re-imputing missing values with median
data['Mileage'].fillna(data['Mileage'].median(), inplace=True)
data['Engine'].fillna(data['Engine'].median(), inplace=True)
data['Power'].fillna(data['Power'].median(), inplace=True)
data['Seats'].fillna(data['Seats'].median(), inplace=True)

# Verify if the missing values are imputed
data.isnull().sum()
```

# B)Remove the units from some of the attributes and only keep the numerical values (for example remove kmpl from “Mileage”, CC from “Engine”, bhp from “Power”, and lakh from“New_price”).

```python
#  'New_Price' contains values in 'Lakh', we will also convert these values to the same scale as 'Price'

data['New_Price'] = data['New_Price'].str.extract('(\d+\.\d+|\d+)').astype(float)

# Display the first few rows to confirm changes happened
data.head()
```

# C)Change the categorical variables (“Fuel_Type” and “Transmission”) into numerical one hot encoded value.

```python
import pandas as pd
# One-hot encoding the categorical variables 'Fuel_Type' and 'Transmission'
data_encoded = pd.get_dummies(data, columns=['Fuel_Type', 'Transmission'])
data_encoded.head()
```

# D)  Create one more feature and add this column to the dataset. For example, you can calculate the current age of the car by subtracting “Year” value from the current year

```python
# Calculating the current age of the car from the 'Year' column
current_year = 2024
data_encoded['Car_Age'] = current_year - data_encoded['Year']

# Display the first few rows to confirm
data_encoded.head()
```

# E) Perform select, filter, rename, mutate, arrange and summarize with group by operations (or their equivalent operations in python) on this dataset

```python
#  Selecting a subset of columns
selected_columns = data_encoded[['Location', 'Year', 'Kilometers_Driven', 'Fuel_Type_Diesel', 'Transmission_Automatic', 'Price', 'Car_Age']]

# Filtering cars older than 10 years
filtered_data = selected_columns[selected_columns['Car_Age'] > 10]

#  Renaming 'Kilometers_Driven' to 'KMs_Driven'
renamed_data = filtered_data.rename(columns={'Kilometers_Driven': 'KMs_Driven'})

# Adding a new column 'Price_Per_Year' as Price divided by Car_Age
renamed_data['Price_Per_Year'] = renamed_data['Price'] / renamed_data['Car_Age']

# Sorting the data by 'Price' in descending order
arranged_data = renamed_data.sort_values(by='Price', ascending=False)

#  Calculating average price and average KMs driven by location
summary = arranged_data.groupby('Location').agg(Avg_Price=('Price', 'mean'), Avg_KMs_Driven=('KMs_Driven', 'mean')).reset_index()

# Displaying the first few rows of the summary data
summary.head()
```
